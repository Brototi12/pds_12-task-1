**#Grade Calculation**

```python
students = [
    {"id": "ST-001", "name": "John Doe", "math_marks": 85, "english_marks": 78, "science_marks": 92},
    {"id": "ST-002", "name": "Jane Smith", "math_marks": 90, "english_marks": 88, "science_marks": 84},
    {"id": "ST-003", "name": "Emily Davis", "math_marks": 72, "english_marks": 68, "science_marks": 75},
    {"id": "ST-004", "name": "Michael Brown", "math_marks": 95, "english_marks": 91, "science_marks": 89},
    {"id": "ST-005", "name": "Sarah Wilson", "math_marks": 65, "english_marks": 60, "science_marks": 58},
    {"id": "ST-006", "name": "David Johnson", "math_marks": 80, "english_marks": 85, "science_marks": 87},
]

def calculate_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    else:
        return 'F'

for student in students:
    total_marks = (student["math_marks"] + student["english_marks"] + student["science_marks"]) / 3
    student["grade"] = calculate_grade(total_marks)
for student in students:
    print(f"ID: {student['id']}, Name: {student['name']}, Grade: {student['grade']}")
```
